## Lab 2. Text Normalization (cleaning)

Depending on the task we have to perform, we can choose one or more ways to clean the text:
- Convert text to lowercase letters
- Remove digits and numbers (or turn them into words)
- Removal of links
- Removal of emoticons ( :) :D) and emojis (üíô üê±)
- Removal of punctuation
- Eliminating stopwords
- Stemming/Lematization
- Tokenization

We will use the `twitter_samples` dataset from `nltk` (https://www.nltk.org/), which contains positive and negative tweets. We will only use positive tweets from now on.


```python
import nltk
nltk.download('twitter_samples')
```

```python
from nltk.corpus  import twitter_samples
tweets = twitter_samples.strings('positive_tweets.json')
tweets = tweets[500:520]
tweets
```

### Transformarea textului √Æn litere mici

```python
tweets_lower = [tweet.lower() for tweet in tweets]
tweets_lower
```

### Removing digits and numbers (or turning them into words)

Removing digits using the `re` regex

```python
import re

tweets_no_digits = [re.sub(r'\d+', '', tweet) for tweet in tweets_lower]
tweets_no_digits
```

Convertirea cifrelor √Æn numere folosind `num2words`

```python
! pip install num2words
```

```python
from num2words import num2words

tweets_num2words = []
for tweet in tweets_lower:
    tweets_num2words.append(' '.join([num2words(word) if word.isdigit() else word for word in tweet.split()]))

tweets_num2words
```

### Eliminarea link-urilor

```python
tweets_no_links = [re.sub(r'http\S+', '', tweet) for tweet in tweets_no_digits]
tweets_no_links
```

### Removal of emoticons ( :) :D) and emojis (üíô üê±)

Removing emoticons using regex from [nltk Twitter Tokenizer](https://github.com/nltk/nltk/blob/develop/nltk/tokenize/casual.py)

```python
emoticon_string = r"""
    (?:
      [<>]?
      [:;=8]                     # eyes
      [\-o\*\']?                 # optional nose
      [\)\]\(\[dDpP/\:\}\{@\|\\] # mouth
      |
      [\)\]\(\[dDpP/\:\}\{@\|\\] # mouth
      [\-o\*\']?                 # optional nose
      [:;=8]                     # eyes
      [<>]?
      |
      </?3                       # heart
    )"""
    
emoticon_re = re.compile(emoticon_string, re.VERBOSE | re.I | re.UNICODE)
tweets_no_emoticons = [re.sub(emoticon_re, '', tweet) for tweet in tweets_no_links]
tweets_no_emoticons
```

Removing emojis using `emoji` [library](https://github.com/carpedm20/emoji)

```python
! pip install emoji
```

`get_emoji_regexp()` returns a regex that encompasses all emojis.

```python
import emoji

emoji_re = emoji.get_emoji_regexp()

tweets_no_emoji = [re.sub(emoji_re,'', tweet) for tweet in tweets_no_emoticons]
tweets_no_emoji
```

### Removal of hashtags and mentions

```python
tweets_no_hashtags = [re.sub(r'#[a-zA-Z0-9_]+','', tweet) for tweet in tweets_no_emoji]
tweets_no_hashtags
```

```python
tweets_no_mentions = [re.sub(r'@[a-zA-Z0-9_]+','', tweet) for tweet in tweets_no_hashtags]
tweets_no_mentions
```

### We remove multiple spaces between words

```python
tweets_no_spaces = [re.sub(r'\s+', ' ', tweet).strip() for tweet in tweets_no_mentions]
tweets_no_spaces
```

### Tokenization

- At sentence level: we can split a text into sentences by punctuation, or using [nltk.sent_tokenize](https://www.nltk.org/api/nltk.tokenize.html)

- At word level: we can split by space, or use [nltk.word_tokenize](https://www.nltk.org/api/nltk.tokenize.html)

Sentence-level tokenization using regex


```python
twitter_sent_tokenized = [re.split('(?<=[.!?])\s+', tweet) for tweet in tweets_no_spaces]
twitter_sent_tokenized
```

Using `nltk.sent_tokenize`

```python
from nltk import sent_tokenize

twitter_sent_tokenized = [nltk.sent_tokenize(tweet) for tweet in tweets_no_spaces]
twitter_sent_tokenized
```

Word-level tokenization by spaces

Word-level tokenization using `nltk.word_tokenize`

## Removal of punctuation

Using regex

```python
tweets_no_punct = [re.sub(r'[^\w\s]', '', tweet) for tweet in tweets_no_spaces]
tweets_no_punct
```

Using `string.punctuation()`

```python
import string

string.punctuation
```

Traducem fiecare semn de punctua»õie √Æn ' '.

```python
tweets_no_punct = [tweet.translate(str.maketrans('', '', string.punctuation)) for tweet in tweets_no_spaces]
tweets_no_punct
```

Word-level tokenization after removing punctuation.

```python
from nltk import word_tokenize

twitter_word_tokenized = [word_tokenize(tweet) for tweet in tweets_no_punct]
twitter_word_tokenized
```

Be careful with cases of the "one, two" type. If you remove the punctuation directly, the two words will be concatenated to obtain a single word "unudoi". An alternative would be to first replace all punctuation characters with spaces, then apply the space contraction method once more.

```python
tweets_no_punct = [re.sub(r'[^\w\s]', ' ', tweet) for tweet in tweets_no_spaces]
tweets_no_punct = [re.sub(r'\s+', ' ', tweet) for tweet in tweets_no_punct]
tweets_no_punct
```

### Eliminating stopwords

![stopwords.jpg](https://user.oc-static.com/upload/2021/01/06/16099626487943_P1C2.png) 

Stopwords are the most used words in the English language and have syntactic and morphological value, but they also have no semantic value.
The first 4 stopwords of the English language in the order of their appearance in the language: the, of, and, to.
Most are pronouns, prepositions or conjunctions.

[The Zipf Law](https://www.youtube.com/watch?v=fCn8zs912OE) states that if we note $x$ = the number of occurrences of "the", then the number of occurrences of "of" will be very close to $\frac{x}{2}$, of "and", of $\frac{x}{3}$, of the nth with $\frac{x}{n} $ (Pareto probability distribution).

## Lemmatization
Match the words to their form in the dictionary.

"Lemmatization is the process where we take individual tokens from a sentence and we try to reduce them to their base form. The process that makes this possible is having a vocabulary and performing morphological analysis to remove inflectional endings. The output of the lemmatization process ( as shown in the figure above) is the lemma or the base form of the word. For instance, a lemmatization process reduces the inflections, "am", "are", and "is", to the base form, "be". Take a look at the figure above for a full example and try to understand what it's doing.

Lemmatization is helpful for normalizing text for text classification tasks or search engines, and a variety of other NLP tasks such as sentiment classification. It is particularly important when dealing with complex languages ‚Äã‚Äãlike Arabic and Spanish." ([source](https://colab.research.google.com/github/dair-ai/notebooks/blob/master/_notebooks/2020-03-19-nlp_basics_tokenization_segmentation.ipynb#scrollTo=dcaLqxPX5CJa))

## Pros:
- the results are existing words in the language

## Cons
- the search process can be expensive. computationally
- the language is in a continuous evolution, and most of the time, the dictionaries do not "keep up" with all the new words that appear or are borrowed

## Stemming
Apply algorithms that use finite automata (remember LFA). Many of them are written in [Snowball](http://snowball.tartarus.org/) language.

"Stemming is just a simpler version of lemmatization where we are interested in stripping the suffix at the end of the word. When stemming we are interested in reducing the inflected or derived word to it's base form. Take a look at the figure above to get some intuition about the process.

Both the stemming and the lemmatization processes involve morphological analysis where the stems and affixes (called the morphemes) are extracted and used to reduce inflections to their base form. For instance, the word cats has two morphemes, cat and s, the cat being the stem and the s being the affix representing plurality." ([source](https://colab.research.google.com/github/dair-ai/notebooks/blob/master/_notebooks/2020-03-19-nlp_basics_tokenization_segmentation.ipynb#scrollTo=dcaLqxPX5CJa))

## Pros:
- it is faster than lemmatization
- adapts more easily to new words appearing in the language

## Cons
- the results are not always words existing in the language

```python
doc = nlp("Apples and oranges are similar boots and hippos aren't")
for word in doc:
    print(word, '=>', word.lemma_)
```

```python
# spaCy doesn't support stemming so for this part we are going to use NLTK, which is another fantastic Python NLP library.
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer(language='english')
doc = 'I prefer not to argue'
for token in doc.split(" "):
    print(token, '=>' , stemmer.stem(token))
```

## Morphological analysis

POS (part of speech) is the part of speech at the macro level.

 - ADJ: adjective, e.g. big, old, green, incomprehensible, first  
 - ADP: adposition, e.g. in, to, during  
 - ADV: adverb, e.g. very, tomorrow, down, where, there  
 - AUX: auxiliary, e.g. is, has (done), will (do), should (do)  
 - CONJ: conjunction, e.g. and, or, but  
 - CCONJ: coordinating conjunction, e.g. and, or, but  
 - DET: determiner, e.g. a, an, the  
 - INTJ: interjection, e.g. psst, ouch, bravo, hello  
 - NOUN: noun, e.g. girl, cat, tree, air, beauty  
 - NUM: numeral, e.g. 1, 2017, one, seventy-seven, IV, MMXIV  
 - PART: particle, e.g. ‚Äôs, not,  
 - PRON: pronoun, e.g I, you, he, she, myself, themselves, somebody  
 - PROPN: proper noun, e.g. Mary, John, London, NATO, HBO  
 - PUNCT: punctuation, e.g. ., (, ), ?  
 - SCONJ: subordinating conjunction, e.g. if, while, that  
 - SYM: symbol, e.g. $, %, ¬ß, ¬©, +, ‚àí, √ó, √∑, =, :), üòù  
 - VERB: verb, e.g. run, runs, running, eat, ate, eating  
 - X: other, e.g. sfpksdpsxmsa  
 - SPACE: space, e.g.


The tag is a fine grained pos, which can be seen as an abbreviation of the description obtained from the morph.

```python
doc = nlp(sentences[0])

descriptions = []
for token in doc:
    desc = [token.text, token.pos_, token.tag_, token.morph]
    descriptions.append(desc)

print(tabulate(descriptions, headers=["word", "pos", "tag", "morph"]))
```

## Lemmatization using POS

We saw in the second lab how we can use both `nltk` and `spacy' to extract the lemmas (dictionary forms) of some words.

In general, lemmatization in `nltk` does not use information about the part of speech, while in `spacy` in the preprocessing pipeline the parts of speech are detected and used in lemmatization. However, we can use the POS in the case of `nltk` as follows:

```python
import nltk
nltk.download('wordnet')
wl = nltk.WordNetLemmatizer()  # instantiem un lemmatizer
```

```python
print(wl.lemmatize("being"))
print(wl.lemmatize("being", pos="n"))
print(wl.lemmatize("being", pos="v"))
```

```python
print(wl.lemmatize("painted"))
print(wl.lemmatize("painted", pos="v"))
print(wl.lemmatize("painted", pos="a"))
```

POS in `nltk' are:
- `n` (noun/noun)
- `v` (verb)
- `a` (adjective)
- `r` (adverb).
