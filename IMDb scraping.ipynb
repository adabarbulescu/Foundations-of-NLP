```python
#importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
# Make a request to the IMDb Top 250 movies page
url = 'https://www.imdb.com/chart/top/'
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the movie links in the page
movie_links = soup.find_all('td', class_='titleColumn')

# For each movie link, get the review page link
for movie_link in movie_links:
    review_link = movie_link.find('a')['href']
    print(review_link)
```
